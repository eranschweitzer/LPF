## Varying $s$
In the following plots $s$ is varied from 0 to 1. The solution is acurate at $s=0$ as expected. However, as $s\to1$ the errors increase.

```matlab
for s = 0:0.1:1
    [v,err,mpc] = resatscale(s,'case300',mpopt);
    titlestr = sprintf('s=%0.1f',s);
    plot_scaling_results(v,err,mpc,titlestr)
end
```
